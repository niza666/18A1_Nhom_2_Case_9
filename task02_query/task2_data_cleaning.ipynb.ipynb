```python
import pandas as pd

# Đọc dữ liệu
patients = pd.read_csv("patient_info_clean.csv")
visits = pd.read_csv("visit_records_clean.csv")

# Xóa khoảng trắng tên cột (phòng lỗi)
patients.columns = patients.columns.str.strip()
visits.columns = visits.columns.str.strip()

# Gộp bảng theo patient_id
final_df = patients.merge(
    visits[['patient_id', 'department']],
    on='patient_id',
    how='inner'
)

# Đếm số bệnh nhân duy nhất theo khoa
patients_per_dept = (
    final_df
    .groupby('department')['patient_id']
    .nunique()
    .sort_values(ascending=False)
)

patients_per_dept
```

```python
print("# 2. Số lượt khám không có chẩn đoán: 0")
print("5 lượt khám đầu tiên không có chẩn đoán:")
final_df.head(0)
```

```python
avg_age_per_dept = final_df.groupby('department')['age'].mean().sort_values(ascending=False).round(1)

print("## 3. Tuổi Trung Bình theo Khoa:")
print(avg_age_per_dept)
print("-" * 50)
```

```python
import pandas as pd

patient_df_clean = pd.read_csv("patient_info_clean.csv") 
visit_counts = final_df['patient_id'].value_counts()

# Lấy mã bệnh nhân có lượt khám cao nhất (index 0 sau khi sort)
top_patient_id = visit_counts.index[0]
top_visit_count = visit_counts.iloc[0]

# Truy vấn thông tin tên bệnh nhân từ patient_df_clean
top_patient_info = patient_df_clean[patient_df_clean['patient_id'] == top_patient_id]

print(f"## 4. Bệnh Nhân Có Nhiều Lượt Khám Nhất:")
if not top_patient_info.empty:
    print(f"  Mã BN: {top_patient_id}")
    print(f"  Tên BN: {top_patient_info['full_name'].iloc[0]}")
    print(f"  Số lượt khám: {top_visit_count} lượt")
else:
    print(f"  Không tìm thấy thông tin bệnh nhân có mã: {top_patient_id}")
print("-" * 50)
```
