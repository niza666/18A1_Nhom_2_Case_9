```python
import pandas as pd
import numpy as np

df = pd.read_csv("visit_records.csv")
df.head()
```

```python
df["patient_id"] = df["patient_id"].str.strip().str.upper()

df["department"] = df["department"].str.strip().str.lower()
dept_map = {
    "nhi": "Nhi",
    "noi": "Noi",
    "no i": "Noi",
    "ngoai": "Ngoai",
    "san": "San"
}
df["department"] = df["department"].map(dept_map)
```

```python
df["diagnosis"] = df["diagnosis"].str.strip().str.lower()

diag_map = {
    "tieu chay": "Tieu Chay",
    "dau bung": "Dau Bung",
    "viem phoi": "Viem Phoi",
    "!@#$": "Chua Ro",
    "": "Chua Ro"
}
df["diagnosis"] = df["diagnosis"].replace(diag_map).fillna("Chua Ro")

df["visit_date"] = pd.to_datetime(df["visit_date"], errors="coerce")
```
