```python
import pandas as pd

patient = pd.read_csv("../data_clean/patient_info_clean.csv")
visit   = pd.read_csv("../data_clean/visit_records_clean.csv")
billing = pd.read_csv("../data_clean/billing_info_clean.csv")

visit_patient_left = pd.merge(
    visit,
    patient,
    on="patient_id",
    how="left"
)

visit_patient_left.head()
```

```python
visit_patient_billing_left = pd.merge(
    visit_patient_left,
    billing,
    on="visit_id",
    how="left"
)

visit_patient_billing_inner = pd.merge(
    visit_patient_left,
    billing,
    on="visit_id",
    how="inner"
)

print(visit_patient_billing_left.shape)
print(visit_patient_billing_inner.shape)
```
